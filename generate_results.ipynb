```python
import torch

model_G = torch.jit.load("./results/G_model256.pth")
```

```python
noise = torch.randn(10, 64, 1, 1)
```

```python
model_G.eval()
generated = model_G(noise).detach()
```

```python
import matplotlib.pyplot as plt
```

```python
from torchvision.utils import save_image, make_grid
```

```python
grid = make_grid(generated, 5)
```

```python
for img in generated:
    img_permute = img.permute(1, 2, 0)    #[height, width, color_channels]
    plt.figure(figsize=(10, 7))
    plt.imshow(img_permute)
    plt.axis("off")
    plt.title('Image', fontsize=14)
    plt.show()
```

```python
def save_graph(y_D, y_G, filepath="./results"):
    # x = np.linspace(0, 10, 100)
    fig, ax = plt.subplots()
    
    ax.plot(y_D, linewidth=2.0, color='green', label='Discriminator Loss')
    ax.plot(y_G, linewidth=2.0, color='red', label='Generator Loss')
    ax.legend()
    # ax.set(xlim=(0, 50))
    
    # plt.show()
    plt.savefig(filepath+"/training_graph.png")
    
```

```python
save_graph([0.00, 0.5, 0.1, 5, 3.0, 2.5],[2.0, 1, 1.5, 0.7, 0.3])
```
